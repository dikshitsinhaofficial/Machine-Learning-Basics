>1. House Price Prediction:-

You are given a dataset containing house features such as area, number of bedrooms, number of
bathrooms, and age of the house. Build a Linear Regression model to predict the price of a house.
Perform necessary data preprocessing and evaluate the model using R² score and RMSE.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# 1. Dataset [cite: 112, 239]
data = {
    'Area': [1100, 1400, 1750, 2100, 2450, 1250, 2050, 1550],
    'Bedrooms': [2, 3, 3, 4, 4, 2, 4, 3],
    'Bathrooms': [1, 2, 2, 3, 3, 1, 2, 2],
    'Age': [12, 6, 3, 9, 2, 14, 4, 18],
    'Price': [160000, 210000, 265000, 310000, 360000, 155000, 295000, 190000]
}
df = pd.DataFrame(data)

# 2. Preprocessing & Splitting [cite: 121, 124, 251]
X = df[['Area', 'Bedrooms', 'Bathrooms', 'Age']]
y = df['Price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# 3. Model Implementation [cite: 128, 252, 253]
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# 4. Evaluation [cite: 131, 255, 256]
print("--- House Price Prediction ---")
print(f"R2 Score: {r2_score(y_test, y_pred):.4f}")
print(f"RMSE: {np.sqrt(mean_squared_error(y_test, y_pred)):,.2f}")
```



>2. Customer Purchase Prediction:-

A company wants to predict whether a customer will purchase a product based on age, gender, and
salary. Implement a classification model (Decision Tree or Logistic Regression) to predict the
purchase decision

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, confusion_matrix

# 1. Dataset [cite: 145, 270]
data = {
    'Age': [28, 42, 33, 21, 58, 31, 44, 52],
    'Gender': ['M', 'F', 'M', 'F', 'M', 'M', 'F', 'F'],
    'Salary': [45000, 72000, 58000, 28000, 115000, 52000, 85000, 105000],
    'Purchased': [0, 1, 0, 0, 1, 0, 1, 1]
}
df = pd.DataFrame(data)

# 2. Preprocessing (Encoding Gender) [cite: 154, 155, 277, 278]
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])
X = df[['Age', 'Gender', 'Salary']]
y = df['Purchased']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# 3. Model Implementation [cite: 161, 282, 283]
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)
y_pred = dt_model.predict(X_test)

# 4. Evaluation [cite: 164, 166, 286, 287]
print("--- Customer Purchase Prediction ---")
print(f"Accuracy: {accuracy_score(y_test, y_pred) * 100:.2f}%")
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```



> 3.Spam Email Detection:-

Given a dataset of emails labeled as spam or not spam, preprocess the data and build a classification
model to detect spam emails.



```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# 1. Dataset [cite: 176, 301]
data = {
    'Email': [
        "Special discount just for you!", "Team meeting at 2 PM.",
        "You won a cash prize!", "Please review the documents.",
        "Limited offer, buy now!", "Let's grab coffee today."
    ],
    'Label': [1, 0, 1, 0, 1, 0] # 1=spam, 0=ham
}
df = pd.DataFrame(data)

# 2. Preprocessing (CountVectorizer) [cite: 191, 193, 318, 319]
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['Email'])
y = df['Label']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

# 3. Model Implementation [cite: 197, 321, 322]
nb_model = MultinomialNB()
nb_model.fit(X_train, y_train)
y_pred = nb_model.predict(X_test)

# 4. Evaluation [cite: 201, 325]
print("--- Spam Email Detection ---")
print(f"Accuracy: {accuracy_score(y_test, y_pred) * 100:.2f}%")
```



> 4.Customer Churn Prediction:-

A telecom company wants to predict whether a customer will leave the service. Use features like
tenure, monthly charges, and contract type to build a Random Forest classification model.



```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# 1. Dataset [cite: 210, 336]
data = {
    'Tenure': [10, 2, 52, 3, 68, 8, 30, 62],
    'MonthlyCharges': [55.0, 75.0, 102.5, 48.0, 112.0, 85.0, 68.0, 98.0],
    'Contract': ['Month-to-month', 'Month-to-month', 'Two year', 'Month-to-month',
                 'Two year', 'Month-to-month', 'One year', 'Two year'],
    'Churn': [0, 1, 0, 1, 0, 1, 0, 0]
}
df = pd.DataFrame(data)

# 2. Preprocessing (One-Hot Encoding) [cite: 221, 345]
df = pd.get_dummies(df, columns=['Contract'], drop_first=True)
X = df.drop('Churn', axis=1)
y = df['Churn']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# 3. Model Implementation [cite: 226, 349, 350]
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
y_pred = rf_model.predict(X_test)

# 4. Evaluation [cite: 230, 231, 353, 354]
print("--- Customer Churn Prediction ---")
print(f"Accuracy: {accuracy_score(y_test, y_pred) * 100:.2f}%")
print("Classification Report:\n", classification_report(y_test, y_pred, zero_division=0))
```
